```python
###

img = get_img(frame_path)[1]
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
h, s, v = cv2.split(hsv)
# show(v)

v_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#v.mean()=smoke detection

#fire
ret, thr = cv2.threshold(v_gray, 180, 255,0)
con, hier = cv2.findContours(thr, cv.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
for cnt in con:
    cont_img = cv2.drawContours(img, [cnt], -1, (255,0,0),1)
    
#smoke
ret, thr_sm = cv2.threshold(v_gray, v_gray.mean(), 255,0)
smoke, hier = cv2.findContours(thr_sm,cv.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

for cnt2 in smoke: 
    cont_img = cv2.drawContours(cont_img,[cnt2], -1, (0,255,0),1 )
    
show(cont_img)
```
